```python
""" 
Aqui é criado um dicionário vazio chamado agenda, ao qual serão adicionados os contatos
desejados através de uma chave e conjunto de valores (dados) respectivos a cada um.
"""
agenda = {}
```

```python
def adicionar():
  """
  A função "adicionar" tem por objetivo fazer a adição de um novo contato desejado pelo
  usuário à agenda. Nela o usuário insere os dados que deseja para o novo contato (nome
  telefone, email, twitter e instagram), e confirma ou não a ação, se sim, o novo contato
  é adicionado à agenda, se não, a adição é cancelada. Nessa função também ocorre a
  incrementação do cR (código de registro, criado e definido na função "menuAgenda") do 
  contato, para que não ocorra repetições das chaves no dicionário.
  
  ***Em todo o programa os inputs de nome, tanto para adicionar quanto para consultar, alterar
  ou remover um contato recebem, posteriormente, o comando .title() para que na hora de rodar
  o programa no menu, a busca pelo nome do contato no dicionário não importe ser digitado em
  letra maiúscula ou minúscula.***
  """

  global cR
  nome = str(input('\nDigite o nome do contato: '))
  telefone = str(input('Digite o número de telefone do contato: '))
  email = str(input('Digite o email do contato: '))
  twitter = str(input('Digite o usuário do Twitter do contato: '))
  instagram = str(input('Digite o usuário do Instagram do contato: '))
    
  confirm = str(input('Confirma a adição do novo contato? Digite (S) para sim ou (N) para não: '))

  if confirm.upper() == 'S':
    cR += 1
    agenda[cR] = [nome.title(), telefone, email, twitter, instagram]
    print('Contato adicionado com sucesso!')
  else:
    print('Adição de contato cancelada!')
```

```python
def consultar():
  """
  A função "consultar" tem por objetivo fazer a consulta, na agenda, dos dados de algum
  contato desejado pelo usuário.
  O usuário faz a consulta pelo nome do contato desejado. A agenda é percorrida pelo
  comando "for", dessa forma, para cada cR e seus respectivos dados, é comparado se 
  o primeiro dado da lista dos dados obtida pelo comando .items() de cada cR corresponde
  ao nome digitado pelo usuário para a busca, se houver correspondência, os dados desse(s)
  determinado(s) cR(s) são mostrados pelo comando .get(), ou seja, são mostrados os valores para
  estas chaves; além disso, a variável "found", que começa com o valor 0 é incrementada em 1 se
  houver alguma corrêspondencia, portanto, se ao final da execução do comando "for" ela permanecer
  com o valor 0, significa não foi encontrado nenhum contato correspondente ao nome inserido.
  """

  contato = str(input('Digite o nome do contato para consulta: '))
  found = 0

  for cR, dados in agenda.items():
    if dados[0] == contato.title():
      found += 1
      print('Dados encontrados para este nome:', agenda.get(cR))

  if found == 0:
    print('Nenhum contato com este nome foi encontrado!')
```

```python
def remover():
  """
  A função "remover" tem por objetivo remover da agenda os dados de algum contato desejado
  pelo usuário, ou melhor, apagar algum contato.
  O usuário digita o nome do contato que deseja remover da agenda, dessa forma o comando
  "for" percorre a agenda comparando o nome com o primeiro dado das listas de dados obtida
  pelo comando .items(). Havendo correspondência, os dados para aquele contato são mostrados
  através do seu cR, com o comando .get(); a variável "found" que começa com o valor 0 é 
  incrementada em 1; e a lista "removerContato" que começa vazia, recebe o valor do cR do 
  contato em que houve correspondência pelo comando .append().
  A depender de quantos contato forem encontrados para o nome digitado, determinados pelo
  valor da variável "found", é executada uma parte específica do código. Se for encontrado
  só 1, o valor do cR que foi para a lista "removerContato" é chamado pela sua índice [0] e 
  atribuído seu valor à variável cR, que é utilizada para remover os dados correspondentes na
  agenda pelo comando .pop() se houver a confirmação do usuário. 
  Se forem encontrados 2 ou mais contatos, o usuário escolhe qual deles deve ser removido
  pela ordem em que forem printados na tela, a variável "indice" recebe a "posição" digitada
  pelo usuário, que -1 se torna o índice correto dentro da lista "removerContato" que contém
  o valor do cR que deve ser removido, dessa forma a variável cR recebe o valor presente nesse
  índice e, com a confirmação do usuário, remove esse contato da agenda pelo comando .pop().
  Por fim, se forem encontrados 0 contatos com o respectivo nome, a variável "found" permanece 
  com valor 0 e é printada a mensagem de que nenhum contato foi encontrado.
  """

  contato = str(input('Digite o nome do contato que deseja remover: '))
  removerContato = []
  found = 0
  
  for cR, dados in agenda.items():
    if dados[0] == contato.title():
      found += 1
      print('Contato encontrado:', agenda.get(cR))
      removerContato.append(cR)
  
  if found == 1:
    cR = removerContato[0]
    confirm = str(input('Deseja removê-lo? Digite (S) para sim ou (N) para não: '))
    if confirm.upper() == 'S':
      agenda.pop(cR)
      print('Contato apagado com sucesso!')
    else:
      print('Ação cancelada.')


  if found >= 2:
    indice = int(input('Qual dos contatos deseja apagar?\nDigite (1) para o primeiro, (2) para o segundo e assim por diante: '))
    cR = removerContato[indice - 1]
    print(agenda[cR])
    confirm = str(input('Confirma a remoção do contato acima? Digite (S) para sim ou (N) para não: '))
    if confirm.upper() == 'S':
      agenda.pop(cR)
      print('Contato apagado com sucesso!')
    else:
      print('Ação cancelada.')


  if found == 0:
    print('Nenhum contato com este nome foi encontrado!')
  
```

```python
def alterar():
  """
  A função alterar tem por objetivo fazer a alteração em algum dos dados de um contato
  indicado pelo usuário digitando o seu nome.
  Utilizando a mesma lógica da função "remover", o usuário digita o nome do contato que
  deseja alterar algum dado, a agenda é percorrida pelo comando "for", e para cada contato 
  encontrado com o nome sendo o primeiro valor da lista de dados de cada cR, é armazenado o
  valor do cR na lista inicialmente vazia "alterarContato", e é incrementada em 1 a variável
  "found", que começa com o valor 0.
  Dentro da função "alterar" é declarada a função "ifDado", que determina qual dado será
  alterado dentro da lista de dados de cada contato a depender do desejo do usuário. A função
  "alterarDado", que é declarada dentro de "ifDado" é a função que faz a alteração do dado em
  si, através dos parâmetros que recebe, sendo estes: o dado a ser alterado (digitado pelo usuário)
  e armazenado na variável dado, e "num", que é determinado pela própria função "ifDado" a depender
  da string recebida como valor para o dado a ser alterado; o novo dado digitado para alterar o
  anterior é armazenado na função "novoDado".
  Por fim, dependendo de qual o valor da variável "found", a função "alterar" executa linhas de
  código diferentes. Se "found" = 1, ou seja foi encontrado só 1 contato com o nome digitado
  pelo usuário, o valor do cR armazenado na lista "alterarContato" se encontra no índice [0],
  e é repassado para a variável "cR", sendo assim chamada a função "ifDado", com a função
  "alterarDado" em seu interior, que recebe o valor do cR e o chama dentro da agenda com o
  índice correspondente ao dado informado pelo usuário para fazer a devida alteração para o 
  valor armazenado na variável "novoDado", se a ação for confirmada pelo usuário.
  Se "found" for igual ou maior que 2, ou seja, foram encontrados mais de 1 contato com o
  nome digitado pelo usuário, o usuário faz a escolha de qual dos contatos deseja alterar
  através da ordem em que são printados na tela. A variável "indice" recebe do usuário, a 
  "posição" do contato na ordem, que -1 se torna o índice correto dentro da lista "alterarContato"
  que contém o valor do cR que deve ser alterado, dessa forma a variável cR recebe o valor presente
  nesse índice e, com a confirmação do usuário, chama a função "ifDado" com a função "alterarDado"
  em seu interior para fazer a devida alteração, se confirmada a ação.
  Se "found" = 0, significa que não foi encontrado nenhum contato com o respectivo nome passado
  pelo usuário.
  """

  contato = str(input('Digite o nome do contato que deseja fazer alguma alteração: '))
  alterarContato = []
  found = 0

  for cR, dados in agenda.items():
    if dados[0] == contato.title():
      found += 1
      print('Dados encontrados para este nome:', agenda.get(cR))
      alterarContato.append(cR)

  def ifDado():
    global dado
    dado = str(input('Qual dado deseja alterar? Nome, Telefone, Email, Twitter ou Instagram? '))
    dado = dado.title()

    def alterarDado(dado, num):
      novoDado = str(input('Digite o novo dado para o contato: '))
      if dado == 'Nome':
        novoDado = novoDado.title()
      confirm = str(input('Confirma a alteração? Digite (S) para sim ou (N) para não: '))
      if confirm.upper() == 'S':
        agenda[cR][num] = novoDado
        print(dado, 'do contato alterado com sucesso!')
      else:
        print('Alteração cancelada.')

    if dado == 'Nome':
      alterarDado(dado, 0)        
    elif dado == 'Telefone':
      alterarDado(dado, 1)
    elif dado == 'Email':
      alterarDado(dado, 2)        
    elif dado == 'Twitter':
      alterarDado(dado, 3)
    elif dado == 'Instagram':
      alterarDado(dado, 4)
  
  if found == 1:
    cR = alterarContato[0]
    confirm = str(input('Deseja alterar algum dado deste contato? Digite (S) para sim ou (N) para não: '))
    if confirm.upper() == 'S':  
      ifDado()
    else:
      print('Alteração cancelada.')

  if found >= 2:
    indice = int(input('Qual dos contatos deseja alterar?\nDigite (1) para o primeiro, (2) para o segundo e assim por diante: '))
    cR = alterarContato[indice - 1]
    print(agenda[cR])
    confirm = str(input('Confirma o contato a ser alterado? Digite (S) para sim ou (N) para não: '))
    if confirm.upper() == 'S':
      ifDado()
    else:
      print('Alteração cancelada.')

  if found == 0:
    print('Nenhum contato com este nome foi encontrado!')
```

```python
def relatorio():
  """
  Essa função tem por objetivo gerar um relatório em formato de tabela com os dados de
  todos os contatos presentes na agenda, sempre atualizada. Para isso foi feita a "estrutura"
  da tabela na mão, para que cada dado caia no seu devido título, o que é possível, fazendo
  uso do comando .format(), dentro do comando "for" para os valores de todas as chaves da agenda,
  que percorre toda a agenda formatando desde as chaves (cR), até o último valor da lista de 
  dados referentes a cada cR, para que sejam encaixados dentro da tabela na posição correta.
  """

  print('\n\t\t\t\t  RELATÓRIO DA AGENDA:\n')
  print('***********************************************************************************************')
  print('|Nro|     Nome     |    Telefone    |       Email        |     Twitter     |     Instagram    |')
  print('***********************************************************************************************')
  for cR in agenda.keys():
    print('| {} |    {}    | {}  |  {}  |     {}     |     {}      |'.format(cR, agenda[cR][0], agenda[cR][1], agenda[cR][2], agenda[cR][3], agenda[cR][4]))
    print('***********************************************************************************************')
```

```python
def salvarContatos():
  """
  Essa função tem por objetivo salvar os dados de cada contato em um arquivo CSV separando
  os dados por vírgula.
  Para isso foi importado o módulo "csv" para o python, e utilizado o método with open 
  para abrir (ou no caso, criar) um arquivo chamado "agenda.csv" como "csvAgenda", com 
  o modo 'w' para a escrita dos dados no arquivo. 
  A a variável writer armazena a função "csv.writer" que pega o arquivo csvAgenda e 
  coloca o delimitador ',' para separar os dados. Depois com o comando "writerows()"
  para a variável "writer", são gravados no arquivo CSV os dados presentes para cada
  lista de dados de cada contato dentro da "agenda.values()".
  """

  import csv

  contatos = agenda.values()
  with open('agenda.csv', 'w') as csvAgenda:
    writer = csv.writer(csvAgenda, delimiter = ',')
    writer.writerows(contatos)
  print('Dados dos contatos salvos com sucesso!')
```

```python
def menuAgenda():
  """
  Essa função tem por objetivo ser o menu da agenda, ou seja, quando é chamada, ela
  executa todas as demais funções do programa, de acordo com o desejo do usuário.
  O menu é composto de 7 opções, do 0 ao 6, todas armazenadas na lista "menu". 
  De acordo com o número que o usuário escolher, uma parte específica do código
  é executada, principalmente chamando uma outra função que faz alguma ação na agenda.
  a variável itemMenu começa com o valor 1, que é para permitir o início do laço de
  repetição do comando "while", posteriormente a variável "itemMenu" passa a receber um 
  número digitado pelo usuário, tendo em vista as especificações printadas na tela
  de qual número vai possibilitar determinada ação na agenda. Por conseguinte,
  se o número digitado como valor para a variável "itemMenu" estiver presente na
  lista "menu", as condicionais seguintes vão determinar qual função será executada,
  a depender do número que foi digitado, se não, será impresso na tela para que seja
  digitado um item válido, e aparecem novamente as opções do menu.
  
  Se dentre os itens válidos do menu, for digitado o número 1, é declarada como
  global a variável cR, que tem seu valor inicial determinado aqui, pois se o dicionário
  da agenda estiver vazio, o seu valor inicial é 0, já se tiver algum valor de cR dentro,
  no caso uma chave com um valor numérico que representa um contato na agenda, o cR
  passa a ter o maior valor dentre essas chaves, para que quando executada a função
  adicionar, essa nova chave seja incrementada em 1 e não se repitam os valores das
  chaves (cR's) no dicionário; além disso, é aqui que se determina quantos contatos
  o usuário deseja adicionar, pois é através do valor que o usuário digita para a
  variável "qtd", que se for maior que 0, o programa vai chamar a função "adicionar"
  quantas vezes for necessário até que esse número seja alcançado, através do comando
  "for".

  Se for digitado o número 2 no menu, é executada a função "consultar"; se 3, a função
  "remover"; se 4, a função "alterar"; se 5, a função "relatório"; se 6, a função 
  "salvarContatos"; e, se 0, é quebrado o laço de repetição do comando while através do
  comando "break" e o usuário é "desconectado" da agenda.

  """
  
  menu = [0, 1, 2, 3, 4, 5, 6]

  itemMenu = 1

  while itemMenu:
    print('\n\n\tAGENDA TELEFÔNICA\n')
    print('1 -> Adicionar novo(s) contato(s).')
    print('2 -> Consultar dados de um contato.')
    print('3 -> Remover um contato.')
    print('4 -> Alterar dado de um contato.')
    print('5 -> Gerar relatório da agenda.')
    print('6 -> Salvar contatos em arquivo CSV.')
    print('0 -> Sair.')

    itemMenu = int(input('\nEscolha um item: '))
      
    if itemMenu in menu:
        
      if itemMenu == 1:
        global cR
        if agenda == {}:
          cR = 0
        else:
          cR = max(agenda.keys())
        qtd = int(input('Quantos contatos deseja adicionar? '))
        if qtd > 0:
          for i in range(qtd):
            adicionar()

      elif itemMenu == 2:
        consultar()
        
      elif itemMenu == 3:
        remover()

      elif itemMenu == 4:
        alterar()
      
      elif itemMenu == 5:
        relatorio()
      
      elif itemMenu == 6:
        salvarContatos()

      elif itemMenu == 0:
        print('Você desconectou da agenda.')
        break
      
    else:
      print('Digite um item válido.')
```
